```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16, seed=123)
if torch.cuda.is_available():
  pipe = pipe.to("cuda")
```

```python
image = pipe("Husky puppy playing on muddy puddle").images[0]
image
```

```python
image = pipe("yellow duck carrying a red flag", num_inference_steps=50).images[0]
image
```
